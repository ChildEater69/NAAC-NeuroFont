## Text to hand Writing Convertion.

```python
from PIL import Image, ImageDraw, ImageFont

# Set the text and font properties
text = "My name is Avdhesh Kumar Sharma"
font_size = 50
font_path = r"C:\Users\Acer pc\Downloads\stork_delivery\Stork Delivery.ttf" # Replace with the path to your desired font file

# Set the canvas size based on the text length and font size
text_width = len(text) * font_size
text_height = font_size

# Create a blank image with a white background
image = Image.new("RGB", (text_width, text_height), "white")
draw = ImageDraw.Draw(image)

# Load the desired font
font = ImageFont.truetype(font_path, font_size)

# Draw the text on the image
draw.text((0, 0), text, font=font, fill="black")

# Save the handwritten image
image.save(r"C:\Users\Acer pc\Downloads\try1.png")
```

## Trying to change color of Text

```python
from PIL import Image, ImageDraw, ImageFont

# Set the text and font properties
text = "My name is Avdhesh Kumar Sharma"
font_size = 50
font_path = r"C:\Users\Acer pc\Downloads\stork_delivery\Stork Delivery.ttf"  # Replace with the path to your desired font file

# Set the canvas size based on the text length and font size
text_width = len(text) * font_size
text_height = font_size

# Set the image width and height for full-page size
image_width = text_width
image_height = text_height * 2

# Create a blank image with a white background
image = Image.new("RGB", (image_width, image_height), "white")
draw = ImageDraw.Draw(image)

# Load the desired font
font = ImageFont.truetype(font_path, font_size)

# Calculate the position to center the text horizontally
text_x = (image_width - text_width) // 2

# Draw the text on the image
draw.text((text_x, 0), text, font=font, fill="blue")

# Save the handwritten image
image.save(r"C:\Users\Acer pc\Downloads\try2.png")
```

## Increasing the size of background 

```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# Set the text and font properties
text = "My name is Avdhesh Kumar Sharma"
font_size = 50
font_path = r"C:\Users\Acer pc\Downloads\stork_delivery\Stork Delivery.ttf"  # Replace with the path to your desired font file

# Create a blank image with a white background
background_color = (255, 255, 255)  # RGB values for white
text_color = (0, 0, 0)  # RGB values for black
image_width = 1920  # Width of the page in pixels
image_height = 1080  # Height of the page in pixels
image = Image.new("RGB", (image_width, image_height), background_color)
draw = ImageDraw.Draw(image)

# Load the desired font
font = ImageFont.truetype(font_path, font_size)

# Calculate the size of the text
text_width, text_height = draw.textsize(text, font=font)

# Calculate the position to center the text horizontally and vertically
text_x = (image_width - text_width) // 2
text_y = (image_height - text_height) // 2

# Draw the text on the image
draw.text((text_x, text_y), text, font=font, fill=text_color)

# Save the handwritten image
image.save(r"C:\Users\Acer pc\Downloads\try3.png")

# Display the image
plt.imshow(image)
plt.axis('off')
plt.show()
```
