```python
import matplotlib.pyplot as plt

# Set the text and font properties
text = "My name is Avdhesh Kumar Sharma"
font_size = 30
font_path = r"C:\Users\Acer pc\Downloads\stork_delivery\Stork Delivery.ttf"  # Replace with the path to your desired handwritten font file

# Create a blank figure and axes
fig, ax = plt.subplots(figsize=(8, 6), dpi=100)

# Set the background color
ax.set_facecolor("white")

# Hide the axes and ticks
ax.axis("off")

# Set the position to top-left corner
text_x = 0
text_y = 1

# Load the desired handwritten font
prop = plt.matplotlib.font_manager.FontProperties(fname=font_path, size=font_size)

# Draw the text on the axes
ax.text(text_x, text_y, text, fontproperties=prop)

# Save the image
plt.savefig(r"C:\Users\Acer pc\Downloads\image2.png", bbox_inches="tight", pad_inches=0)

print("Image saved as 'image.png'. Please open the file using an image viewer to see the result.")
```
